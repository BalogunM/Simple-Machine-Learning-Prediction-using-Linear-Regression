```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
data = pd.read_csv('happyscore_income.csv')
income = data['avg_income']
happy =data['happyScore']
plt.xlabel('income')
plt.ylabel('happy')
plt.scatter(income,happy)
data.sort_values('country', inplace=True)


from pandas.plotting import scatter_matrix  ##some intro to data visualization on how to check for correlation between attributes

attributes = ["avg_income", "happyScore", "GDP",
              "income_inequality"]
scatter_matrix(data[attributes], figsize=(12, 8))
```

```python
X = np.c_[data['avg_income']]
y = np.c_[data['happyScore']]
data.plot(kind='scatter',x='avg_income',y='happyScore')
plt.show()
```

```python
import sklearn
lin_reg_model = sklearn.linear_model.LinearRegression()
#lin_reg_model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=2)  ##this line is for k_Nearest Neighbor regression
lin_reg_model.fit(X,y)

#from sklearn.linear_model import LinearRegression

#lin_reg = LinearRegression()
#lin_reg.fit(X, y)
```
